```python
"""1.Social Media – Post Engagement Analyzer:
Create a function to analyze engagement on a social media post.
Expected Output Format
Total Likes: 1050
Post Status: Viral Post"""
#function to calculate likes
def PostLikes(likes):
  total=sum(likes)
  if total>=1000:
    status= "Viral Post"
  else:
    status= "Normal Engagement"
  print(f"Total likes:{total}")
  print(f"Post status:{status}")

likes=[100,200,500,700]
PostLikes(likes)
```

```python
"""3.Agriculture – Rainfall Adequacy Checker:
Create a function that analyzes seasonal rainfall data.
Rules
Calculate average rainfall
Average ≥ required level → Adequate Rainfall
Otherwise → Inadequate Rainfall
Expected Output Format
Average Rainfall: 72
Rainfall Status: Adequate Rainfall"""
#function tocalculate rainfall
def RainfallChecker(rainfall,required_level):
  average=sum(rainfall)/len(rainfall)
  if average >=required_level:
    status="Adequate Rainfall"
  else:
    status="Inadequate Rainfall"
  print(f"Average Rainfall:{average}")
  print(f"Rainfall Status:{status}")
#user input
rainfall_input=input("Enter rainfall:")
rainfall = list(map(int, rainfall_input.split()))
required_level=int(input("Enter required level:"))
RainfallChecker(rainfall,required_level)
```

```python
"""4.Social Media – Duplicate Account Detection:
Create a function to detect duplicate usernames in a system.
Rules
If duplicates exist → Flag accounts
Else → Accounts are unique
Expected Output Format
Duplicate Accounts Found: Yes"""
#function to detect duplicate account
def dupliacteAccounts(username):
  unique_name=set(username)
  if len(unique_name)!=len(username):
    print("Duplicate Account Found:Yes")
  else:
    print("Duplicate Account Found:No")
#user input
user_input=input("Enter username")
username=user_input.split()
dupliacteAccounts(username)
```

```python
"""6.Agriculture – Premium Crop Price Filter:
Create a function that filters crop prices above a premium threshold.
Rules
Price > ₹2000 → Premium Crop
Else → Regular Crop
Expected Output Format
Premium Crops: [2500, 3200]"""
#function to filter the crop price
def CropPrice(prices):
  premium_crops = []
  for price in prices:
    if price > 2000:
      premium_crops.append(price)
  print(f"Premium Crops: {premium_crops}")

# User input
price_input = input("Enter crop prices : ")
prices = list(map(int, price_input.split()))
CropPrice(prices)
```

```python
"""7.System Monitoring – Application Health Checker:
Create a function to evaluate application health based on error count.
Rules
Errors = 0 → Healthy
Errors ≤ 5 → Minor Issues
Errors > 5 → Critical Issues
Expected Output Format
Error Count: 7
System Status: Critical Issues"""
#function to evaluate application health based on error count
def HealthChecker(errors):
  if errors ==0:
    status="Healthy"
  elif errors<=5:
    status="Minor Issues"
  else:
    status="Critical Issues"
  print(f"Error Count:{errors}")
  print(f"System Status:{status}")
#user input
errors=int(input("Enter error count"))
HealthChecker(errors)
```

```python
"""8.Banking – Daily Transaction Limit Checker:
Create a function to validate daily transaction amount.
Rules
Daily limit = ₹50,000
Amount within limit → Approved
Amount exceeds limit → Rejected
Expected Output Format
Transaction Amount: 60000
Transaction Status: Rejected"""
#function to validate daily transaction amount
def DailyTransaction(amount):
  Daily_limit=50000
  if amount<=Daily_limit:
    status="Approved"
  else:
    status="Rejected"
  print(f"Transaction Amount:{amount}")
  print(f"Transaction status:{status}")
#user input
amount=int(input("Enter amount:"))
DailyTransaction(amount)
```

```python
"""9.E-Learning – Student Attendance Eligibility System:
Create a function to calculate attendance percentage and check exam eligibility.
Rules
Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible
Expected Output Format
Attendance Percentage: 80.0
Exam Eligibility: Eligible"""
#function to calculate attendance percentage and check exam eligibility
def AttendanceCalculation(attendance):
  percentage=sum(attendance)/len(attendance)*100
  if percentage>=75:
    status="Eligible"
  else:
    status="Not Eligible"
  print(f"Attendance percentage:{percentage}")
  print(f"Exam Eligibilty:{status}")
#user input
attendance_input=input("Enter attendance(1 for Present, 0 for Absent):")
attendance=list(map(int,attendance_input.split()))
AttendanceCalculation(attendance)
```

```python
"""10.Smart Electricity Bill Analyzer:
Create a function that calculates the electricity bill and classifies usage level based on units consumed.
Billing Rules
First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit
Usage Classification
Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage
Use return values (bill and usage status)"""
#function that calculates the electricity bill and classifies usage level based on units consumed
def ElectricalBill(units):
  bill=0
  if units<=100:
    bill=units*3
  elif units<=200:
    bill=(100*3)+((units-100)*5)
  else:
    bill = (100*3)+(100*5)+((units-200)*7)
  if bill<500:
    status="Low usage"
  elif bill<=1500:
    status="Moderate usage"
  else:
    status="High Usage"
  print(f"Total bill:₹{bill}")
  print(f"Usage Level:{status}")
#user input
units=int(input("Enter units:"))
ElectricalBill(units)
```
