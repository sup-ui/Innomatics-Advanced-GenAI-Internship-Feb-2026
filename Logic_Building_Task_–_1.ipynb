```python
"""1.User Login Check:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"
Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"""

username="admin"
password="1234"
#user input
input_username=input("Enter username:")
input_password=input("Enter password:")
#logic
if input_username == username and input_password==password:
  print("Login Succesfully")
else:
  print("Invalid Credentials")
```

```python
"""2.Pass / Fail Analyzer:
Analyze student results.
marks = [45, 78, 90, 33, 60]
Requirements:
A student passes if marks â‰¥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly"""

marks=[45,78,90,33,60]
pass_count=0
fail_count=0
#logic
for i in marks:
  if i >=50:
    pass_count=pass_count+1
  else:
    fail_count=fail_count+1
#print statements
print("Total number of students: ",len(marks))
print("Total number of pass students: ",pass_count)
print("Total number of failed students: ",fail_count)
```

```python
"""3. Simple Data Cleaner:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]
Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list"""

names=[" Alice ", "bob", "CHARLIE "]
#creating an empty list
list_names=[]
#logic
for i in names:
  clean_names=i.strip().lower()
  list_names.append(clean_names)
#print
print("Clean names:",list_names)
```

```python
"""4. Message Length Analyzer:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]
Requirements:
Print the length of each message
Flag messages longer than 10 characters"""

messages=["Hi", "Welcome to the platform", "OK"]
#logic
for i in messages:
  msg_length=len(i)
  print("Message:",i)
  print("Length:",msg_length)
  #to see if message is more than 10 character
  if(msg_length)>10:
    print("More than 10 character")
```

```python
"""5. Error Message Detector:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
Requirements:
Count the number of "ERROR" entries"""

logs=["INFO", "ERROR", "WARNING", "ERROR"]
sum=0
#logic
for i in logs:
  if i=="ERROR":
    sum=sum+1
print("Count of error entries:",sum)
```
